```python
import csv
from pytube import YouTube
from googleapiclient.discovery import build

# Read video URLs from the CSV file
api_key = 'AIzaSyA8Bqd1F0qr_Wou3ub641AstJdA_hdPBdk'

# Create a YouTube Data API client
youtube = build('youtube', 'v3', developerKey=api_key)

video_urls = []
with open('D:\mlproject\data.csv','r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        video_urls.append(row['video url'])
        
video_data = []

# Loop through each video URL and get information
for idx, video_url in enumerate(video_urls, start=1):
    try:
        yt_video = YouTube(video_url)
        # Get video ID from the URL using pytube
        video_id = yt_video.video_id
        
        # Use YouTube Data API to get likes and comments counts
        video_response = youtube.videos().list(
            part='statistics',
            id=video_id
        ).execute()
        
        video_stats = video_response['items'][0]['statistics']
        likes_count = int(video_stats.get('likeCount', 0))
        comments_count = int(video_stats.get('commentCount', 0))
        
        video_data.append({
            'Index': idx,
            'Video Url': video_url,
            'Video Title': yt_video.title,
            'Video Views': yt_video.views,
            'Video Likes': likes_count,
            'Video Comments': comments_count,
            'Video Author': yt_video.author,
            'Duration In Seconds': yt_video.length
        })

    except Exception as e:
        print(f"Error processing {video_url}: {e}")
        print("\n")
        
output_csv_file = 'Youtube_data.csv'
with open(output_csv_file, 'w', newline='', encoding='utf-8') as csv_file:
    fieldnames = ['Index', 'Video Url', 'Video Title', 'Video Views', 'Video Likes', 'Video Comments', 'Video Author', 'Duration In Seconds']
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()

    for data in video_data:
        # Encode each field, ignoring problematic characters
        encoded_data = {k: v.encode('utf-8', 'ignore').decode('utf-8') if isinstance(v, str) else v for k, v in data.items()}
        csv_writer.writerow(encoded_data)

print(f"Video data saved to {output_csv_file}")
```
